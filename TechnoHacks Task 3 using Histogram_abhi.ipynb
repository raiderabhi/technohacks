# <center>Iris Flower Visualization</center>

The goal of this project is to build a classification model that classify flower on the basis of its features. Here's a brief description of the columns:

+ sepal_length: This column represents the length of the sepal of the iris flower. Sepals are the green, leaf-like structures that protect the flower bud.

+ sepal_width: This column represents the width of the sepal of the iris flower.

+ petal_length: This column represents the length of the petal of the iris flower. Petals are the colorful, often delicate, leaf-like parts of the flower that are responsible for attracting pollinators.

+ petal_width: This column represents the width of the petal of the iris flower.

+ species: This column likely contains information about the species or variety of iris flower being measured. Common species in datasets like this include "setosa," "versicolor," and "virginica."

## Data Visualization

**Let's make seperate list of categorical and numerical variables**

```python
cat_var = ['species']
num_var = ['sepal_length', 'sepal_width','petal_length','petal_width']
```

**Visualize the survival outcomes of individuals**

```python
ax = sns.countplot(x=df['species'], palette='RdBu')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
for column in num_var:
    plt.figure(figsize=(14,5))
    plt.subplot(1,2,1)
    ax = sns.boxplot(df[column])

    plt.subplot(1,2,2)
    ax = sns.distplot(df[column])
    plt.show()
```

**Pairplot to visualize relationships between variables**

```python
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
```

**Violin plot to visualize distribution and density by species.**

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.violinplot(x="species", y="petal_length", data=df)
plt.subplot(2, 2, 2)
sns.violinplot(x="species", y="petal_width", data=df)
plt.subplot(2, 2, 3)
sns.violinplot(x="species", y="sepal_width", data=df)
plt.subplot(2, 2, 4)
sns.violinplot(x="species", y="sepal_length", data=df)
plt.tight_layout()
```

```python
# Correlation heatmap
data = df.copy()
data.drop("species", axis=1, inplace=True)
correlation = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
